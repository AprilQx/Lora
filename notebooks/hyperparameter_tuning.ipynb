```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Create figure and axis
fig, ax = plt.subplots(figsize=(14, 8))

# Define colors
colors = {
    "base": "#AED6F1",  # Light blue
    "hyper": "#D5F5E3",  # Light green
    "context": "#FADBD8",  # Light red
    "final": "#E8DAEF",  # Light purple
    "eval": "#FCF3CF",   # Light yellow
    "arrow": "#566573"   # Dark gray
}

# Create boxes for different components with proper spacing
components = [
    # Base model section
    {"name": "Base Model", "position": (0.05, 0.8), "width": 0.15, "height": 0.1, 
     "color": colors["base"], "section": "Base"},
    {"name": "Qwen2.5-0.5B-Instruct", "position": (0.05, 0.65), "width": 0.15, "height": 0.1, 
     "color": colors["base"], "section": "Base"},
    
    # Hyperparameter experiments section
    {"name": "Hyperparameter\nExperiments", "position": (0.3, 0.8), "width": 0.15, "height": 0.1, 
     "color": colors["hyper"], "section": "Hyper"},
    {"name": "Learning Rate\n(1e-5, 5e-5, 1e-4)", "position": (0.25, 0.65), "width": 0.15, "height": 0.1, 
     "color": colors["hyper"], "section": "Hyper"},
    {"name": "LoRA Rank\n(2, 4, 8)", "position": (0.25, 0.5), "width": 0.15, "height": 0.1, 
     "color": colors["hyper"], "section": "Hyper"},
    {"name": "Precision\n(2, 3 decimal places)", "position": (0.25, 0.35), "width": 0.15, "height": 0.1, 
     "color": colors["hyper"], "section": "Hyper"},
    {"name": "Best Config:\nLR=1e-4, Rank=8", "position": (0.45, 0.65), "width": 0.15, "height": 0.1, 
     "color": colors["hyper"], "section": "Hyper"},
    
    # Context length experiments section
    {"name": "Context Length\nExperiments", "position": (0.55, 0.8), "width": 0.15, "height": 0.1, 
     "color": colors["context"], "section": "Context"},
    {"name": "128 Tokens", "position": (0.55, 0.65), "width": 0.15, "height": 0.08, 
     "color": colors["context"], "section": "Context"},
    {"name": "512 Tokens", "position": (0.55, 0.55), "width": 0.15, "height": 0.08, 
     "color": colors["context"], "section": "Context"},
    {"name": "768 Tokens", "position": (0.55, 0.45), "width": 0.15, "height": 0.08, 
     "color": colors["context"], "section": "Context"},
    {"name": "Best Length:\n512 Tokens", "position": (0.75, 0.65), "width": 0.15, "height": 0.1, 
     "color": colors["context"], "section": "Context"},
    
    # Final training section
    {"name": "Final Training", "position": (0.8, 0.35), "width": 0.15, "height": 0.1, 
     "color": colors["final"], "section": "Final"},
    {"name": "LR=1e-4\nRank=8\nContext=512", "position": (0.8, 0.2), "width": 0.15, "height": 0.1, 
     "color": colors["final"], "section": "Final"},
    
    # Evaluation section
    {"name": "Evaluation", "position": (0.55, 0.2), "width": 0.15, "height": 0.1, 
     "color": colors["eval"], "section": "Eval"},
    {"name": "MAE, RMSE\nFLOP Usage", "position": (0.3, 0.2), "width": 0.15, "height": 0.1, 
     "color": colors["eval"], "section": "Eval"}
]

# Draw boxes
for comp in components:
    rect = patches.Rectangle(comp["position"], comp["width"], comp["height"], 
                             linewidth=1, edgecolor='black', facecolor=comp["color"])
    ax.add_patch(rect)
    ax.text(comp["position"][0] + comp["width"]/2, comp["position"][1] + comp["height"]/2, 
            comp["name"], ha='center', va='center', fontsize=9)

# Add arrows
# Base to Hyper
ax.arrow(0.21, 0.7, 0.03, 0, head_width=0.015, head_length=0.01, fc=colors["arrow"], ec=colors["arrow"])
# Hyper to Context
ax.arrow(0.61, 0.7, 0.03, 0, head_width=0.015, head_length=0.01, fc=colors["arrow"], ec=colors["arrow"])
# Context to Final (diagonal)
ax.arrow(0.8, 0.65, 0, -0.19, head_width=0.015, head_length=0.01, fc=colors["arrow"], ec=colors["arrow"])
# Final to Eval
ax.arrow(0.79, 0.25, -0.08, 0, head_width=0.015, head_length=0.01, fc=colors["arrow"], ec=colors["arrow"])
# Eval to left
ax.arrow(0.29, 0.25, -0.08, 0, head_width=0.015, head_length=0.01, fc=colors["arrow"], ec=colors["arrow"])

# Add section labels
ax.text(0.125, 0.92, "1. BASE MODEL", fontsize=12, weight='bold', ha='center')
ax.text(0.425, 0.92, "2. HYPERPARAMETER SEARCH", fontsize=12, weight='bold', ha='center')
ax.text(0.675, 0.92, "3. CONTEXT LENGTH STUDY", fontsize=12, weight='bold', ha='center')
ax.text(0.875, 0.5, "4. FINAL MODEL", fontsize=12, weight='bold', ha='center', rotation=90)
ax.text(0.425, 0.08, "5. EVALUATION", fontsize=12, weight='bold', ha='center')

# Remove axis
ax.axis('off')
plt.tight_layout()

# Save the figure

plt.show()

print("Methodology diagram created successfully!")
```
